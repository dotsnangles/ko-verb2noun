```python
import pandas as pd
from module.regularize import sentence_to_noun_verb
```

```python
sheet_1 = 'data/eshc 인과관계 학습용(sample)_rev2.xlsx'
sheet_2 = 'data/port 인과관계 학습용(sample)_rev4.xlsx'
data_1 = pd.read_excel(sheet_1, names=[f'Col {i}' for i in range(10)])
data_2 = pd.read_excel(sheet_2, names=[f'Col {i}' for i in range(9)])
```

```python
ce_chains_1 = data_1.iloc[:, 1:]
ce_chains_2 = data_2.iloc[:, 1:]
```

```python
ce_chains_1_morphed = ce_chains_1.applymap(lambda x: sentence_to_noun_verb(x) if pd.isna(x) == False else x)
ce_chains_2_morphed = ce_chains_2.applymap(lambda x: sentence_to_noun_verb(x) if pd.isna(x) == False else x)
```

```python
original = ce_chains_1.values.tolist() + ce_chains_2.values.tolist()
morphed = ce_chains_1_morphed.values.tolist() + ce_chains_2_morphed.values.tolist()

for ori, mor in zip(original, morphed):
    print(ori)
    print(mor)
    print()
```
